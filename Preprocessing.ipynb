```python
def percent_reviews(rating):
    '''
    Function that returns a list of the percent each rating received
    Input: <int> rating
    Output: list of percentages for that rating
           
    '''
    percents = [data['Reviews'][i][0][rating - 1] for i in range(len(data))]
        
    return percents
```

```python
def get_meta_data(element):
    '''
    Function that grabs the metadata based on the location in the split list and the length of the split list
    Input: <int> negative element placement
    Output: <list> list of the meta data
    '''
        
    meta = []
    for i in range(len(split_df)):
        split_elt = split_df['Reviews'][i].split('\n')
        length = len(split_elt)
        
        #rating for lens 6 and 8, unknown for lens 5 and 7
        if element == -3:
            if (length == 7) or (length == 5):
                meta.append('unknown')
            else:
                meta.append(split_elt[element])
         
        #review
        elif element == -4:
            if (length == 7) or (length == 5):
                meta.append(split_elt[-3])
            else:
                meta.append(split_elt[element])
             
        #date
        elif element == -5:
            if (length == 7) or (length == 5):
                meta.append(split_elt[-4])
            else:
                meta.append(split_elt[element])
        
        #length taken for len 8 or 7, unknown otherwise
        elif element == -6:
            if  len(split_elt) == 6 or len(split_elt) == 5:
                meta.append('unknown')
            elif len(split_df) == 7:
                meta.append(split_elt[-5])
            else:
                meta.append(split_elt[element])
        else:
            meta.append(split_elt[element])
            
    return meta
```

```python
split_df['raw_review'] = get_meta_data(-4)
split_df['review_date'] = get_meta_data(-5)
split_df['user_rating'] = get_meta_data(-3)
split_df['num_helpful_votes'] = get_meta_data(-1)
split_df['length_of_use'] = get_meta_data(-6)
```
